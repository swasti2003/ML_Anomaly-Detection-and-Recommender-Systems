Here are what the columns represent:

credit.policy: 
1 if the customer meets the credit underwriting criteria

purpose: 
The purpose of the loan (takes values "creditcard", "debtconsolidation", "educational", "majorpurchase", "smallbusiness", and "all_other").

int.rate: 
The interest rate of the loan, as a proportion (a rate of 11% would be stored as 0.11). Borrowers to be more risky are assigned higher interest rates.

installment: 
The monthly installments owed by the borrower if the loan is funded.

log.annual.inc: 
The natural log of the self-reported annual income of the borrower.

dti: 
The debt-to-income ratio of the borrower (amount of debt divided by annual income).

fico: 
The FICO credit score of the borrower.

days.with.cr.line: 
The number of days the borrower has had a credit line.

revol.bal: 
The borrower's revolving balance (amount unpaid at the end of the credit card billing cycle).

revol.util: 
The borrower's revolving line utilization rate (the amount of the credit line used relative to total credit available).

inq.last.6mths: 
The borrower's number of inquiries by creditors in the last 6 months.

delinq.2yrs: 
The number of times the borrower had been 30+ days past due on a payment in the past 2 years.

pub.rec: 
The borrower's number of derogatory public records (bankruptcy filings, tax liens, or judgments).

```python
 #applying filter to the dataset to see the users relevant to the clustering process
 # users with fico above 600 and dti below or equal to 20 are considered as viable borrowers
 loan_data['filter'] = (loan_data['fico'] > 600) & (loan_data['dti'] <= 20)
```

```python
loan_data.head()
```

```python
loan_data.loc[loan_data['filter'] == True]
```

```python
df = loan_data.loc[loan_data['filter'] == True]
df.head()
```

```python
#looking at the shape of final data set
df.shape
```

```python
conditions = [
    (df['fico'] > 600) & (df['fico'] <= 650),
    (df['fico'] > 650) & (df['fico'] <= 700),
    (df['fico'] > 700) & (df['fico'] <= 750),
    (df['fico'] > 750) & (df['fico'] <= 800),
    (df['fico'] > 800) & (df['fico'] <= 850)]
```

```python
values = ['tier_5', 'tier_4', 'tier_3', 'tier_2', 'tier_1']
```

```python
df['tier'] = np.select(conditions, values)
```

```python
df.head()
```

```python
df['user_id'] = range(1, len(df) + 1)
```

```python
df['fico_1'] = df['fico']
```

```python
df_new = df.groupby(by = ['fico'])['fico_1'].count().reset_index().rename(columns={'fico_1':'total'})
```

```python
df_new.head()
```

```python
df_final = df.merge(df_new, left_on = 'fico', right_on = 'fico', how = 'left')
df_final.head()
```

```python


#df_final = df_final.drop('purpose',1)
df_final = df_final.drop('tier',1)
```

```python
## First lets create a Pivot matrix

pivot_table=df_final.pivot_table(index='user_id',columns='fico',values='total').fillna(0)
pivot_table
```

```python
from scipy.sparse import csr_matrix

df_matrix = csr_matrix(pivot_table.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(df_matrix)
```

```python
query_index = np.random.choice(pivot_table.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(pivot_table.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 5)
```

```python
print(indices)
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(pivot_table.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, pivot_table.index[indices.flatten()[i]], distances.flatten()[i]))
```

```python
# First let's create a dataset called X
X = np.array(df_final)
 # create train and test
tpercent = 0.9
tsize = int(np.floor(tpercent * len(df_final)))
df_final_train = X[:tsize]
df_final_test = X[tsize:]
#len of train and test
lentrain = len(df_final_train)
lentest = len(df_final_test)
```

```python
# KNN classifier to find similar products
print(lentrain)
print(lentest)
```

```python
df_final.head()
```

```python
neighbor = NearestNeighbors(n_neighbors=3, algorithm='ball_tree').fit(df_final_train)

# Let's find the k-neighbors of each point in object X. To do that we call the kneighbors() function on object X.
distances, indices = neighbor.kneighbors(df_final_train)
```

```python
df_final.columns
```

```python
#find most related products
for i in range(lentest):
    a = neighbor.kneighbors([df_final_test[i]])
    related_product_list = a[1]

    first_related_fico = [user_id[0] for user_id in related_product_list]
    first_related_fico = str(first_related_fico).strip('[]')
    first_related_fico = int(first_related_fico)
    second_related_fico = [user_id[1] for user_id in related_product_list]
    second_related_fico = str(second_related_fico).strip('[]')
    second_related_fico = int(second_related_fico)
    
    print ("Based on product reviews, for ", df_final["user_id"][lentrain + i] ," average rating is ",df_final["fico"][lentrain + i])
    print ("The first similar product is ", df_final["user_id"][first_related_fico] ," average rating is ",df_final["fico"][first_related_fico])
    print ("The second similar product is ", df_final["user_id"][second_related_fico] ," average rating is ",df_final["fico"][second_related_fico])
    print ("--------------------------------------------------------")
```
